```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
v= np.array([3,-1])
v2= np.array([2,4])

r= v + v2

plt.quiver(0,0,v[0],v[1],color = 'r',lw=0.5,scale_units ='xy',scale=1,angles ='xy',label = 'First vector')
plt.quiver(v[0],v[1],v2[0],v2[1],color = 'g',lw=0.5,scale_units ='xy',scale=1,angles ='xy',label='Secoond Vector')
plt.quiver(0,0,r[0],r[1],color = 'blue',lw=0.5,scale_units ='xy',scale=1,angles ='xy',label='Resultant Vector')
plt.xlim(-1,6)
plt.ylim(-2,5)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Vector addition')
plt.axhline(0,color='Blue',lw=0.5)
plt.axvline(0,color='Blue',lw=0.5)
plt.grid()
plt.legend()
plt.show()
```

```python
v= np.array([2,3])

c = 2
r = c*v
plt.figure(figsize=(6,6))
plt.quiver(0,0,v[0],v[1],color = 'r',lw=0.5,scale_units ='xy',scale=1,angles='xy',label = 'Vector v')
plt.quiver(0,0,r[0],r[1],color = 'r',lw=0.5,scale_units ='xy',scale=1,angles='xy',label = 'Vector v')

plt.xlim(-6,10)
plt.ylim(-6,10)
plt.axhline(0,color = 'b',lw=0.5)
plt.axvline(0,color = 'b',lw=0.5)
plt.grid()
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Vector v and its scalar multiple r')
plt.legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x_values = np.linspace(-4,8,100)

y1 = 2 * x_values
y2 = (x_values + 3) / 2

inter1 , inter2 = 1,2

plt.figure(figsize=(8,6))
plt.xlim(-4,4)
plt.ylim(-4,4)
plt.plot(x_values,y1,label = '$2x - y = 0$',color = 'r')
plt.plot(x_values,y2,label = '$2x - y = 0$',color = 'r')
plt.plot(inter1,inter2,'ro')
plt.text(inter1+0.5,inter2+0.5,f'({inter1},{inter2})')
plt.axhline(0,color='b',lw=0.5)
plt.axvline(0,color='b',lw=0.5)
plt.title('Graph of 2x - y = 0')
plt.xlabel('x')
plt.ylabel('Y')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x_values = np.linspace(-10,10,100)
y_values = np.linspace(-10,10,100)

X,Y = np.meshgrid(x_values,y_values)
Z1 = 1 - X + 2 * Y

Z2 = 2*X + Y-2

Z3 = 3- 3 * X - Y

fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(111,projection = '3d')

ax.plot_surface(X,Y,Z1,color = 'b',alpha=0.5,rstride=100,cstride=100)
ax.plot_surface(X,Y,Z2,color = 'b',alpha=0.5,rstride=100,cstride=100)
ax.plot_surface(X,Y,Z3,color = 'b',alpha=0.5,rstride=100,cstride=100)
```

```python
import numpy as np
import matplotlib.pyplot as plt

a_col1 = np.array([2,-1])
a_col2 = np.array([-1,2])*2

b = np.array([0,3])
fig,ax= plt.subplots(figsize=(8,6))

ax.quiver(0,0,a_col1[0],a_col1[1],scale=1,scale_units='xy',angles='xy',label =f'$A[:,0] = [2,-1]$',color ='red')
ax.quiver(a_col1[0],a_col1[1],a_col2[0],a_col2[1],scale=1,scale_units='xy',angles='xy',label =f'$A[:,0] = [-1,2]$',color='blue')
ax.quiver(0,0,b[0],b[1],scale=1,scale_units='xy',angles='xy',label =f'Resultant Vector',color='green')

ax.set_xlim(-3,3)
ax.set_ylim(-3,3)
ax.set_title('Column picture of the system: Linear Combination of column vectors')
ax.axhline(0,lw = 0.6,color = 'black')
ax.axvline(0,lw = 0.6,color='black')
ax.set_xlabel('X-Axis')
ax.set_ylabel('Y-Axis')
plt.legend()
```
