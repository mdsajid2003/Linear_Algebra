```python
def cal_dis(point):
    dim = len(point)
    dis  = np.sqrt(np.sum(np.square(point)))
    return dis, dim
point_1d = np.array([6])
distance_1, dimen_1 = cal_dis(point_1d)

point_2d = np.array([5,9])
distance_2, dimen_2 = cal_dis(point_2d)

point_3d = np.array([3,6,5])
distance_3, dimen_3 = cal_dis(point_3d)

#display the results
print(f"1d space: distance from origin :{distance_1} in {dimen_1} dimension")
print(f"1d space: distance from origin :{distance_2} in {dimen_2} dimension")
print(f"1d space: distance from origin :{distance_3} in {dimen_3} dimension")
```

```python
def cal_d(point1,point2):
    dim = len(point1)
    distance = np.linalg.norm(np.array(point1)- np.array(point2))
    return distance,dim
point_1d_a = np.array([6])
point_1d_b = np.array([7])
distance_1, dimen_1 = cal_d(point_1d_a,point_1d_b)

point_2d_a = np.array([5,9])
point_2d_b = np.array([7,8])
distance_2, dimen_2 = cal_d(point_2d_a,point_2d_b)

point_3d_a = np.array([3,6,5])
point_3d_b = np.array([7,8,9])
distance_3, dimen_3 = cal_d(point_3d_a,point_3d_b)


#display the results
print(f"1d space: distance from point{point_1d_a} to  point{point_1d_b} is {distance_1} in {dimen_1} dimension")
print(f"2d space: distance from point{point_2d_a} to  point{point_2d_b} is {distance_2} in {dimen_2} dimension")
print(f"3d space: distance from point{point_3d_a} to  point{point_3d_b} is {distance_3} in {dimen_3} dimension")

fig, ax  = plt.subplots(nrows=1,ncols=2,figsize=(10,6))
ax[0].axhline(0,color='red',lw = 0.5)
ax[0].axvline(0,color='red',lw = 0.5)
ax[0].quiver(0,0,*point_2d_a,color='blue',scale_units = 'xy',angles='xy',scale=1)
ax[0].quiver(0,0,*point_2d_b,color='green',scale_units = 'xy',angles='xy',scale=1)
ax[0].scatter(*zip(point_2d_a,point_2d_b),color='red')
ax[0].plot(*zip(point_2d_a,point_2d_b),color='green',ls='--')
ax[0].set_ylabel('Y-axis')
ax[0].set_xlabel('X-axis')


ax= fig.add_subplot(122,projection = '3d')
ax.quiver(0,0,0,*point_3d_a,color = 'blue')
ax.quiver(0,0,0,*point_3d_b,color = 'blue')
ax.scatter(*zip(point_3d_a,point_3d_b),color='red')
ax.plot(*zip(point_3d_a,point_3d_b),color='green',ls='--')
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Arc

import numpy as np

np.random.seed(42)
def plot_v(ax,v1,v2,angles):
    dot_p = np.dot(v1,v2)

    v1_mag = np.linalg.norm(v1)
    v2_mag = np.linalg.norm(v2)

    cos_angle = dot_p / (v1_mag*v2_mag)


    ax.quiver(0,0,v1[0],v1[1],angles='xy',scale_units ='xy',scale=1,color='red')
    ax.quiver(0,0,v2[0],v2[1],angles='xy',scale_units ='xy',scale=1,color='blue')

    angle_rad = np.deg2rad(angles)

    arc= Arc((0,0),0.5,0.5,theta1=0,theta2=angles,color='green')
    ax.add_patch(arc)

    ax.set_xlim(-2,2)
    ax.set_ylim(-2,2)
    ax.grid(True)
    ax.axhline(y=0,color = 'k',ls='--',lw=0.5)
    ax.axvline(x=0,color = 'k',ls='--',lw=0.5)
    ax.set_aspect('equal')

    ax.text(0.05,-1.8,f'Dot Product: {dot_p: .2f}')
    ax.text(0.05,-1.6,f'Cosine of Angle: {cos_angle: .2f}')
    ax.text(0.05,-1.4,f'Angle: {angles} degrees')

    ax.set_title(f'Vector Dot Product at {angles}')
    plt.legend()
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(2,2,figsize=(12,12))

v1 = np.array([1,0])
angles = [0,45,90,180]
for ax, angle in zip([ax1,ax2,ax3,ax4],angles):
    angle_rad = np.deg2rad(angle)
    v2 = np.array([np.cos(angle_rad),np.sin(angle_rad)])
    plot_v(ax,v1,v2,angle)
plt.tight_layout()
# plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import CountVectorizer
import seaborn as sns
movies = {
    "The Dark Knight": {
        "description": "superhero action crime thriller batman fights joker in gotham city",
        "genres": [1, 1, 1, 0, 0]  # [Action, Crime, Thriller, Comedy, Romance]
    },
    "Heat": {
        "description": "crime thriller police detective chases bank robbers in los angeles",
        "genres": [0, 1, 1, 0, 0]  # [Action, Crime, Thriller, Comedy, Romance]
    },
    "The Hangover": {
        "description": "comedy friends bachelor party funny adventure in las vegas",
        "genres": [0, 0, 0, 1, 0]  # [Action, Crime, Thriller, Comedy, Romance]
    },
    "Notting Hill": {
        "description": "romantic comedy love story bookstore owner meets famous actress",
        "genres": [0, 0, 0, 1, 1]  # [Action, Crime, Thriller, Comedy, Romance]
    },
    "Die Hard": {
        "description": "action thriller police officer fights terrorists in building",
        "genres": [1, 0, 1, 0, 0]  # [Action, Crime, Thriller, Comedy, Romance]
    }
}
vectonizer = CountVectorizer(stop_words='english')
descriptions = [movies[title]['description']for title in movies]
description_vectors = vectonizer.fit_transform(descriptions).toarray()
genre_vectors = np.array([movies[title]['genres'] for title in movies])


def calculate_similarities(vectors):
    normalized_vectors = vectors / np.linalg.norm(vectors,axis=1)[:,np.newaxis]
    similarities = np.dot(normalized_vectors, normalized_vectors.T)
    return similarities


desc_similarities = calculate_similarities(description_vectors)
genre_similarities = calculate_similarities(genre_vectors)

fig, (ax1,ax2) = plt.subplots(1,2,figsize=(15,6))

sns.heatmap(desc_similarities,annot=True,cmap='YlOrRd',xticklabels=movies.keys(),yticklabels=movies.keys(),ax=ax1)
ax1.set_title('Movie Similarities based on Descriptions\n(Dot Product)')

sns.heatmap(genre_similarities,annot=True,cmap='YlOrRd',xticklabels=movies.keys(),yticklabels=movies.keys(),ax=ax2)
ax2.set_title('Movie Similarities based on Descriptions\n(Dot Product)')

plt.tight_layout()
plt.show()


```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_vector(a,b):
    b_norm = np.linalg.norm(b)
    proj = (np.dot(a,b) / (b_norm**2)) * b
    fig, ax = plt.subplots(figsize = (10,10))
    ax.set_aspect('equal')

    ax.grid(True)

    origin = np.array([0,0])

    ax.quiver(*origin,*a,angles = 'xy',scale_units = 'xy',scale =1,color = 'blue',label='Vector a',width = 0.01)
    ax.quiver(*origin,*b,angles = 'xy',scale_units = 'xy',scale =1,color = 'red',label='Vector a',width = 0.01)
    ax.quiver(*origin,*proj,angles = 'xy',scale_units = 'xy',scale =1,color = 'green',label='Vector a',width = 0.01)
    ax.plot([a[0],proj[0]],[a[1],proj[1]],'k--',lw =1)
    max_val = max(np.max(np.abs(a)),np.max(np.abs(b))) * 1.2
    ax.set_xlim(-max_val,max_val)
    ax.set_ylim(-max_val,max_val)

    ax.set_xlabel('x')
    ax.set_xlabel('y')
    ax.set_title('Vector projection')
    ax.legend()

    proj_magnitude = np.linalg.norm(proj)
    text = (f'Vector a = [{a[0]:.1f},{a[1]:.1f}]\n'
            f'Vector b = [{b[0]:.1f},{b[1]:.1f}]\n'
            f' projection mag = {proj_magnitude:.2f}')
    ax.text(0.02, 0.95, text, transform = ax.transAxes,verticalalignment ='top',bbox = dict(boxstyle='round',facecolor = 'white',alpha=0.5))
    return fig, ax

a = np.array([2,4])
b= np.array([8,3])

fig,ax = plot_vector(a,b)
plt.show()
```

Divides the feature space in 2 boundaries

```python
def plot_2d_line(w0,w1,w2,x1_range=(-10,10),title = "2D Line Visualization"):
    fig ,ax = plt.subplots(figsize=(10,8))

    x1 = np.linspace(x1_range[0],x1_range[1],100)

    x2 = -(w0+w1*x1)/w2

    ax.plot(x1,x2,'-b',label = f'{w1}x1 + {w2}x2 + {w0} = 0',lw = 2)

    mid_point = len(x1)//2

    mid_x1 = x1[mid_point]
    mid_x2 = x2[mid_point]

    normal = np.array([w1,w2])

    normal = normal / np.linalg.norm(normal)

    arrow_length = 2
    ax.arrow(mid_x1,mid_x2,arrow_length*normal[0],arrow_length*normal[1],head_width =0.5,head_length = 0.6,fc = 'g',ec = 'g',label = 'Positive Region')
    ax.arrow(mid_x1,mid_x2,-arrow_length*normal[0],-arrow_length*normal[1],head_width =0.5,head_length = 0.6,fc = 'r',ec = 'r',label = 'Negative Region')
    ax.grid(True,ls = '--',alpha = 0.5)

    ax.set_aspect('equal')

    ax.set_xlabel('x1')
    ax.set_ylabel('x2')
    ax.set_title(title)
    ax.legend()
    equation_text = f'Equation: {w1}x1 + {w2}x2 + {w0} = 0'
    slope_text = f'Slope {-w1/w2:.2f}'
    intercept_text = f'x2Intercept {w0/w2:.2f}'
    ax.text(0.02,0.9,f'{equation_text}\{slope_text}\n{intercept_text}', transform=ax.transAxes,verticalalignment = 'top',bbox = dict(boxstyle='round',facecolor = 'white',alpha=0.8))
    margin = 2
    ax.set_xlim(x1_range[0]-margin,x1_range[1]+margin)
    ax.set_ylim(min(x2)-margin,max(x2)+margin)

    return fig,ax
w0,w1,w2 = 3,2,
fig1,ax1 = plot_2d_line(w0,w1,w2,title="Example line 2x1+4x2+3 = 0")

plt.show()
```
